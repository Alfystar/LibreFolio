Funzione di ricerca ETF per nome,ticker, isin o wkn:

```python
# 1. Definisci SOLO le colonne "normali" (escludi l'indice 'isin' da qui)
cols_only = ['name', 'ticker', 'wkn']
valore_cercato = 'world'

# 2. Cerca nelle colonne (Vettoriale)
# Questo crea una serie di True/False solo per le colonne
mask_cols = df_all[cols_only].astype(str).agg(' '.join, axis=1).str.contains(valore_cercato, case=False)

# 3. Cerca nell'Indice (Direttamente, senza reset_index)
# df.index accede all'indice senza copiarlo.
# .to_series() crea una vista leggera per usare i metodi .str
mask_index = df_all.index.to_series().astype(str).str.contains(valore_cercato, case=False)

# 4. Unisci i risultati con OR logico (|)
# True se trovato nelle colonne OPPURE nell'indice
filtro_finale = mask_cols | mask_index

# 5. Applica il filtro
risultato = df_all[filtro_finale]
print(filtro_finale)
risultato
```

```python
import json
from websocket import create_connection


def get_realtime_snapshot(isin):
    """
    Si connette al WebSocket, scarica il primo messaggio (stato attuale)
    e chiude immediatamente la connessione.
    """
    url = f"wss://api.mobile.stock-data-subscriptions.justetf.com?subscription=trend&parameters=isins:{isin}/currency:EUR/language:en"

    # Headers per simulare un browser ed evitare blocchi
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)",
        "Origin": "https://www.justetf.com"
        }

    try:
        # 1. Apre la connessione (timeout breve per non restare appesi)
        # create_connection è sincrono: aspetta finché non è connesso
        ws = create_connection(url, header=headers, timeout=5)

        # 2. Legge il primo messaggio disponibile
        # Di solito le API "subscription" mandano subito lo stato corrente appena ti connetti
        result = ws.recv()

        # 3. Chiude subito la connessione per non lasciare socket aperti
        ws.close()

        # 4. Converte la stringa JSON in un dizionario Python
        if result:
            return json.loads(result)
        else:
            return None

    except Exception as e:
        print(f"Errore durante il recupero: {e}")
        return None


# --- ESEMPIO DI UTILIZZO ---
isin_code = "IE00B3RBWM25"
data = get_realtime_snapshot(isin_code)

if data:
    # Visualizza il JSON formattato bene
    print(json.dumps(data, indent=2))

    # Esempio: Se volessi estrarre un valore specifico (la struttura dipende dalla risposta)
    # print(f"Ultimo prezzo: {data.get('valore', 'N/A')}")
```
